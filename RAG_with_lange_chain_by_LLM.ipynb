```python
from transformers import LlamaTokenizer, LlamaForCausalLM, GenerationConfig, pipeline
import torch

from langchain.llms import HuggingFacePipeline
model_id = "chavinlo/alpaca-native"

tokenizer = LlamaTokenizer.from_pretrained(model_id)

base_model = LlamaForCausalLM.from_pretrained(
    model_id,
    load_in_8bit=True,
    device_map='auto',
)
```

```python
from langchain.llms import HuggingFacePipeline

pipe = pipeline("text-generation",
                model=base_model,
                tokenizer=tokenizer,
                max_length=256,
                temperature=0.7,
                top_p=0.95,
                repetition_penalty=1.2)

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
template = """Below is an instruction that describes a task. Write a response that appropriately completes the request.

### Instruction:
Suggest 2 ways to lose my weight. Set order for each way.

Answer:"""

print(llm(template))
```

```python
from langchain.document_loaders import WikipediaLoader

query_1 = "Cristiano Ronaldo"
query_2 = "Lionel Messi"
query_3 = "Karim Benzema"

########################################
docs_1 = WikipediaLoader(query=query_1, load_max_docs=2,
                         doc_content_chars_max=20_000).load()

text_document_1 = docs_1[0].page_content

########################################
docs_2 = WikipediaLoader(query=query_2, load_max_docs=2,
                         doc_content_chars_max=20_000).load()

text_document_2 = docs_2[0].page_content
########################################
docs_3 = WikipediaLoader(query=query_3, load_max_docs=2,
                         doc_content_chars_max=20_000).load()

text_document_3 = docs_3[0].page_content
```

```python
documents = [ text_document_1, text_document_2,text_document_3]
metadatas = [ {"document":query_1}, {"document":query_2} ,{"document":query_3}]
```

```python
documents[0]
```

```python
from langchain.text_splitter import NLTKTextSplitter
import nltk
nltk.download('punkt')

text_splitter = NLTKTextSplitter(chunk_size=300, chunk_overlap=50)
```

```python
tokens_chunks = text_splitter.create_documents(documents, metadatas=metadatas)
```

```python
len(tokens_chunks)
```

```python
tokens_chunks[5]
```

```python
from langchain.embeddings import SentenceTransformerEmbeddings

model_name = "sentence-transformers/all-MiniLM-L6-v2"

embedding_llm = SentenceTransformerEmbeddings(model_name=model_name)
```

```python
# source data
docs_text = [ chunk.page_content for chunk in tokens_chunks ]
docs_embeddings = embedding_llm.embed_documents(docs_text)
```

```python
query_text = "who is Portuguese professional footballer?"
query_embedding = embedding_llm.embed_query(query_text)
```

```python

from langchain_chroma import Chroma

# store text documents as vectors
save_to_dir = "/content/drive/MyDrive/Wiki_chroma_db"

docs_ids = list( range( len(tokens_chunks) ) )
docs_ids = [ str(d) for d in docs_ids ]

vector_db = Chroma.from_documents(
                                tokens_chunks,
                                embedding_llm,
                                persist_directory=save_to_dir,
                                ids=docs_ids
                            )
```

```python
# search for most similar document to a query
query_text ="who is Portuguese professional footballer?"

similar_docs = vector_db.similarity_search(
                                      query_text,
                                      k=5,
                                      filter={'document': 'Cristiano Ronaldo'}
                                    )
```

```python
#stuff chain
from langchain import PromptTemplate
from langchain.chains.question_answering import load_qa_chain
qna_template = "\n".join([
    "Answer the next questionu using the provided context.",
    "If the answer is not contained in the context, say 'NO ANSWER IS AVAILABLE'",
    "### Context:",
    "{context}",
    "",
    "### Question:",
    "{question}",
    "",
    "### Answer :",
])

qna_prompt = PromptTemplate(
    template=qna_template,
    input_variables=['context', 'question'],
    verbose=True
)

stuff_chain = load_qa_chain(llm, chain_type="stuff", prompt=qna_prompt)
```

```python
question = "Who achive five UEFA Champions Leagues?"

similar_docs_LangChain = vector_db.similarity_search(question, k=3)

print(len(similar_docs_LangChain))
```

```python
answer = stuff_chain(
    {
        "input_documents": similar_docs_LangChain,
        "question": question
    },
    return_only_outputs=False,
)

answer
```
